Here we investigate the `College` dataset, stored in `College.csv`

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

First, read the data in and examine it

```python
college = pd.read_csv("College.csv")
college.head()
```

The first column is just the name of the college - set this as the index with the `names` header. 

```python
# college.columns[] = "names"
college.set_index(college.columns[0], inplace=True)
college.index.name = "Names"
college.head()
```

Look at the data, using pd.DataFrame.describe


```python
college.describe()
```

Plot a scatterplot matrix of the first 10 columns of the matrix

```python
sns.pairplot(college[college.columns[:10]])
# alternatively:
# sns.pairplot(college.iloc[:,:10])
```

```python
# Pairplot silently ignores the "Private" field as it is a categorical variable. We could cast it to a bool
# but let's instead use it to colour the scatterplots
sns.pairplot(college[college.columns[:10]], hue="Private" , plot_kws=dict(alpha=0.25))
```

Side by side boxplots of `Outstate` vs `Private`

```python
sns.boxplot(data=college, x="Outstate", y="Private")
# violinplot probably nicer here
```

Create a new qualitative variable, called `Elite`, by binning the `Top10perc` variable - if the proportion of students coming from the top 10% of their high school exceeds 50%, Elite is "Yes"

```python
# For interest
college.Top10perc.plot.hist(title="Distribution of the proportion of top-decile high school\n students per college ",
                            xlim=(0,100))
# Could use seaborns prettier plotting facilities
```

```python
college["Elite"] = np.where(college['Top10perc']>=50, 'Yes', 'No')
```

How many elite universities are there?

```python
college["Elite"].value_counts()
```

Box plot of `Outstate` vs `Elite`

```python
sns.boxplot(data=college, x="Outstate", y="Elite")
```

Plot some histograms of the quantitative variables (Already did that in the pairplot)

```python
columnsToPlot = ['Apps', 'Accept', 'Enroll', 'Top10perc',
                 'F.Undergrad', 'Outstate', 'Room.Board',
                 'PhD', 'Terminal', 'S.F.Ratio', 'Expend',
                 'Grad.Rate']
fig, axes = plt.subplots(nrows=len(columnsToPlot), figsize=(10,20))
for column, ax in zip(columnsToPlot,axes):
    sns.distplot(college[column], hist=False, ax=ax , kde_kws={"shade": True, "cut": 0})
plt.tight_layout()
```
